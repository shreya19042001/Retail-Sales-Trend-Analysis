```python
from pyspark.sql import SparkSession
from pyspark.sql import functions as F
import matplotlib.pyplot as plt

# Step 1: Create Spark Session
spark = SparkSession.builder.appName("SalesAnalysis").getOrCreate()

# Step 2: Read CSV
df = spark.read.csv("cleaned_sales_data.csv", header=True, inferSchema=True)

# Step 3: Calculate Revenue (if not already in file)
df = df.withColumn("Revenue", F.col("Quantity") * F.col("UnitPrice"))

# Step 4: Extract Month
df = df.withColumn("Month", F.date_format("InvoiceDate", "yyyy-MM"))

# Step 5: Monthly Sales Summary
monthly_sales = df.groupBy("Month").agg(F.round(F.sum("Revenue"), 2).alias("Total_Sales"))
monthly_sales.show()

```
